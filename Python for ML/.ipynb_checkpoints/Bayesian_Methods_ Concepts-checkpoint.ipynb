## Baye's Theorem

#### P(A|B) = P(A) * P(B|A) / P(B)

Lets use this in Machine Learning, to create a spam classifier.

Example : How would we express the probability of an email being spam if it contains the "free" word

#### P(spam | free) = P(spam) * P(freee | spam) / P(free)

## Naive Bayes (The Easy Way)

We will cheat by using sklearn.naive_bayes to train a spam classifier! Most of the code is just loading our training data into pandas DataFrame that we can play with.

```python
import os
import io
import numpy
from pandas import DataFrame
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

def readFile(path):
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            path = os.path.join(root,filename)
            
            inBody = False
            lines = []
            f = io.open(path, 'r', encoding='latin1')
            
            for line in f:
                if inBody:
                    lines.append(line)
                elif line == '\n':
                    inBody = True
                    
            f.close()
            
            message = '\n'.join(lines)
            yield path, message
            
            
def dataFrameDir(path, classification):
    rows = []
    index = []
    
    for filename,message in readFile(path):
        rows.append({'message':message,'class':classification})
        index.append(filename)
        
    return DataFrame(rows, index=index)

data = DataFrame({'message':[],'class':[]})

data = data.append(dataFrameDir('/Users/Jarvis/Desktop/machine_learning/Python for ML/spam','spam'))
data = data.append(dataFrameDir('/Users/Jarvis/Desktop/machine_learning/Python for ML/ham','ham'))
```

Let's have a look at that DataFrame

```python
data
```

Now we will use countvectorizer to split up each message into its list of words, and throw that into a MultinomialNB classifier. Call fit() and we've got a trained spam filter ready to go! Its just that easy.

```python
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(data['message'].values)

classifier = MultinomialNB()
targets = data['class'].values

classifier.fit(counts,targets)
```

Let's try it out

```python
examples = ['Free IPhone Now !!!','Hii Sandy How about a game of volley tomorrow']
example_counts = vectorizer.transform(examples)
predictions = classifier.predict(example_counts)
predictions
```
